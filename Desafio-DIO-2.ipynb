## Objetivo:

Crie uma função que recebe como parâmetro a quantidade de vitórias e derrotas de um jogador,
depois disso retorne o resultado para uma variável, o saldo de Rankeadas deve ser feito através do calculo (vitórias - derrotas)

Se vitórias for menor do que 10 = Ferro
Se vitórias for entre 11 e 20 = Bronze
Se vitórias for entre 21 e 50 = Prata
Se vitórias for entre 51 e 80 = Ouro
Se vitórias for entre 81 e 90 = Diamante
Se vitórias for entre 91 e 100= Lendário
Se vitórias for maior ou igual a 101 = Imortal

## Saída

Ao final deve se exibir uma mensagem:
"O Herói tem de saldo de **{saldoVitorias}** está no nível de **{nivel}**"


```python
def Contasaldo(Quant_vitorias, Quant_derrotas):
    saldoVitorias = Quant_vitorias - Quant_derrotas
    return saldoVitorias
def nivelJogador(Quant_vitorias):
    if Quant_vitorias < 10:
        nivel = "Ferro"
    elif 11 <= Quant_vitorias <= 20:
        nivel = "Bronze"
    elif 21 <= Quant_vitorias <= 50:
        nivel = "Prata"
    elif 51 <= Quant_vitorias <= 80:
        nivel = "Ouro"
    elif 81 <= Quant_vitorias <= 90:
        nivel = "Diamante"
    elif 91 <= Quant_vitorias <= 100:
        nivel = "Lendário"
    else:
        nivel = "Imortal"
    return nivel

def main():
    Quant_vitorias = int(input("Escreva a quantidade de vitórias que o jogador possui: "))
    Quant_derrotas = int(input("Escreva a quantidade de derrotas que o jogador possui: "))
    saldoVitorias = Contasaldo(Quant_vitorias, Quant_derrotas)
    elo = nivelJogador(Quant_vitorias)

    print(f"O jogador tem um saldo de {saldoVitorias} e está no nível {elo}")




if __name__ == "__main__":
    main()
```
